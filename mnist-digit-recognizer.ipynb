```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
```

```python
df_train = pd.read_csv('../input/digit-recognizer/train.csv') # train data
df_test = pd.read_csv('../input/digit-recognizer/test.csv') # test data
```

```python
df_train.head()
# It is a collection of pixels of the images 
# label is a number represented in an image i.e actual output
```

```python
img_array = np.array(df_train.iloc[3,1:]).reshape(28,28)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img_array)
```

```python
df_train.shape
```

```python
x_train = df_train.iloc[:,1:] 
y_train = df_train['label'] 
```

```python
x_train = x_train / 255 
# maximum size of pixel is 255 that's why for scaling we can directly divide it by 255
# else we can use other scaling method
```

```python
#output ranges from 0 - 9 so convert it into array format
from tensorflow.keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train,num_classes=10)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# creating a model with input shape
model = Sequential()
model.add(Dense(64 , activation = 'relu', input_shape = (784,)))
model.add(Dense(32 , activation = 'relu'))
model.add(Dense(32 , activation = 'relu'))
model.add(Dense(10 , activation = 'softmax'))
```

```python
model.compile(optimizer='adam',loss = 'categorical_crossentropy', metrics=['accuracy'])
```

```python
# Now out model is ready to train
model.fit(x_train, y_train, epochs=50, batch_size = 64, workers=10)
```

```python
# Saving the model So that we can use it later for prediction
model.save('MNIST_digit_prediction.h5')
```

```python
# Now our model is ready we can predict the output
test_img = np.array(df_test.iloc[4,:]).reshape(28,28)
plt.imshow(test_img)
```
